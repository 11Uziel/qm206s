```python
# Función: Calcular la molaridad de una solución
def calcular_molaridad(gramos, volumen_litros, peso_molecular):
    """
    Calcula la molaridad de una solución.

    Parámetros:
        gramos (float): Masa del soluto en gramos.
        volumen_litros (float): Volumen de la solución en litros.
        peso_molecular (float): Peso molecular del soluto en g/mol.

    Retorna:
        float: Molaridad de la solución.
    """
    moles = gramos / peso_molecular
    molaridad = moles / volumen_litros
    return molaridad

# Ejemplo:
print("Molaridad:", calcular_molaridad(10, 0.5, 58.44))  # Salida esperada: 0.342 M


# Función: Calcular el pH de una solución de ácido fuerte
def calcular_ph(concentracion_h):
    """
    Calcula el pH de una solución de ácido fuerte.

    Parámetros:
        concentracion_h (float): Concentración de iones H+ en mol/L.

    Retorna:
        float: pH de la solución.
    """
    import math
    return -math.log10(concentracion_h)

# Ejemplo:
print("pH:", calcular_ph(0.001))  # Salida esperada: 3.0


# Función: Calcular la masa molar de un gas ideal
def masa_molar_gas(gramos, volumen_ml, temperatura_c, presion_atm):
    """
    Calcula la masa molar de un gas ideal.

    Parámetros:
        gramos (float): Masa del gas en gramos.
        volumen_ml (float): Volumen del gas en mililitros.
        temperatura_c (float): Temperatura en grados Celsius.
        presion_atm (float): Presión en atmósferas.

    Retorna:
        float: Masa molar del gas en g/mol.
    """
    R = 0.0821  # Constante de los gases ideales en L·atm/(mol·K)
    volumen_litros = volumen_ml / 1000
    temperatura_k = temperatura_c + 273.15
    n = (presion_atm * volumen_litros) / (R * temperatura_k)
    masa_molar = gramos / n
    return masa_molar

# Ejemplo:
print("Masa molar del gas:", masa_molar_gas(1, 1000, 25, 1))  # Salida esperada: ~24.04 g/mol


# Función: Preparar un buffer
def preparar_buffer(ph_deseado, ka, volumen_ml, concentracion_acido):
    """
    Calcula las concentraciones del ácido y la base conjugada para preparar un buffer.

    Parámetros:
        ph_deseado (float): pH deseado del buffer.
        ka (float): Constante de acidez del ácido.
        volumen_ml (float): Volumen total del buffer en mililitros.
        concentracion_acido (float): Concentración del ácido en mol/L.

    Retorna:
        tuple: Concentración del ácido y de la base conjugada (en mol/L).
    """
    import math
    h_conjugada = 10 ** (ph_deseado - math.log10(ka))
    concentracion_base = concentracion_acido * h_conjugada
    return concentracion_acido, concentracion_base

# Ejemplo:
acido, base = preparar_buffer(4.75, 1.8e-5, 1000, 0.1)
print("Concentración ácido:", acido, "Concentración base:", base)  # Salida esperada: Ácido ~0.1 M, Base ~0.056 M


# Función: Calcular la energía liberada en una reacción química
def energia_liberada(enthalpia_productos, enthalpia_reactivos):
    """
    Calcula la energía liberada en una reacción química.

    Parámetros:
        enthalpia_productos (float): Entalpía de los productos en kJ/mol.
        enthalpia_reactivos (float): Entalpía de los reactivos en kJ/mol.

    Retorna:
        float: Energía liberada en la reacción (kJ/mol).
    """
    energia = enthalpia_reactivos - enthalpia_productos
    return energia

# Ejemplo:
print("Energía liberada:", energia_liberada(-110, 50))  # Salida esperada: 160 kJ/mol


# Función: Calcular el rendimiento porcentual
def rendimiento_porcentual(masa_teorica, masa_experimental):
    """
    Calcula el rendimiento porcentual de una reacción química.

    Parámetros:
        masa_teorica (float): Masa teórica esperada en gramos.
        masa_experimental (float): Masa experimental obtenida en gramos.

    Retorna:
        float: Rendimiento porcentual.
    """
    rendimiento = (masa_experimental / masa_teorica) * 100
    return rendimiento

# Ejemplo:
print("Rendimiento porcentual:", rendimiento_porcentual(50, 45))  # Salida esperada: 90.0%


# Función: Calcular el calor específico
def calor_especifico(masa, delta_t, calor):
    """
    Calcula el calor específico de un metal.

    Parámetros:
        masa (float): Masa del metal en gramos.
        delta_t (float): Cambio de temperatura en °C.
        calor (float): Calor absorbido o liberado en calorías.

    Retorna:
        float: Calor específico en cal/(g·°C).
    """
    c = calor / (masa * delta_t)
    return c

# Ejemplo:
print("Calor específico:", calor_especifico(50, 25, 500))  # Salida esperada: 0.4 cal/(g·°C)


# Función: Calcular la presión osmótica
def presion_osmotica(concentracion_g_l, temperatura_c, peso_molecular):
    """
    Calcula la presión osmótica a 25 °C.

    Parámetros:
        concentracion_g_l (float): Concentración de la solución en g/L.
        temperatura_c (float): Temperatura en grados Celsius.
        peso_molecular (float): Peso molecular del soluto en g/mol.

    Retorna:
        float: Presión osmótica en atmósferas.
    """
    R = 0.0821
    temperatura_k = temperatura_c + 273.15
    concentracion_molar = concentracion_g_l / peso_molecular
    pi = concentracion_molar * R * temperatura_k
    return pi

# Ejemplo:
print("Presión osmótica:", presion_osmotica(100, 25, 58.44))  # Salida esperada: ~42.56 atm


# Función: Calcular el producto de solubilidad (Kps)
def calcular_kps(solubilidad):
    """
    Calcula el producto de solubilidad (Kps) de una sal insoluble.

    Parámetros:
        solubilidad (float): Solubilidad de la sal en mol/L.

    Retorna:
        float: Producto de solubilidad (Kps).
    """
    kps = solubilidad ** 2  # Para sales 1:1
    return kps

# Ejemplo:
print("Producto de solubilidad:", calcular_kps(0.01))  # Salida esperada: 1e-4
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Función para graficar pH vs. concentración de iones H+
def graficar_ph_concentracion():
    """
    Genera una gráfica de pH en función de la concentración de iones H+.
    """
    # Rango de concentraciones (de 1e-8 a 1 M)
    concentraciones = np.logspace(-8, 0, 100)

    # Cálculo del pH para cada concentración
    ph = -np.log10(concentraciones)

    # Crear la gráfica
    plt.figure(figsize=(8, 5))
    plt.plot(concentraciones, ph, label="pH vs. [H⁺]", color="blue")
    plt.xscale("log")  # Escala logarítmica para la concentración
    plt.xlabel("Concentración de H⁺ (mol/L)")
    plt.ylabel("pH")
    plt.title("Relación entre pH y concentración de H⁺")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.legend()
    plt.show()

# Llamar a la función para generar la gráfica
graficar_ph_concentracion()
```
